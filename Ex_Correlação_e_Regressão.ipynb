# Caso 1

1. Faça a correlação de Pearson. Qual interpretação do resultado?
2. Plote o gráfico.
3. Faça a regressão. Tem sentido? Calcule y para x = 110
4. Calcule o erro quadrático médio

```python
from scipy.stats import pearsonr

# Dados
x = [89, 43, 36, 36, 95, 10, 66, 34, 38, 20, 26, 29, 48, 64, 6, 5, 36, 66, 72, 40]
y = [21, 46, 3, 35, 67, 95, 53, 72, 58, 10, 26, 34, 90, 33, 38, 20, 56, 2, 47, 15]

# Cálculo da correlação de Pearson
corr, _ = pearsonr(x, y)
print(f"Correlação de Pearson: {corr}")
```

```python
import matplotlib.pyplot as plt

# Gráfico de dispersão
plt.scatter(x, y)
plt.title('Gráfico de Dispersão de x e y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
import numpy as np

# Regressão linear
coef = np.polyfit(x, y, 1)  # Ajuste de 1º grau (linear)
polynomial = np.poly1d(coef)

# Calcular y para x = 110
x_new = 110
y_pred = polynomial(x_new)
print(f"Valor de y para x = {x_new}: {y_pred}")

# Plotar a linha de regressão
plt.scatter(x, y)
plt.plot(x, polynomial(x), color='red')
plt.title('Gráfico de Dispersão com Linha de Regressão')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

Não faz sentido, pois a correlação é muito próximo de 0, fazendo com que ao alterar o X, não afeta o Y.

```python
# Calcular as previsões
y_pred_all = polynomial(x)

# Erro Quadrático Médio
mse = np.mean((np.array(y) - y_pred_all) ** 2)
print(f"Erro Quadrático Médio: {mse}")
```

# Caso 2

1. Faça a correlação de pearson. Qual interpretação do resultado?
2. Plote o gráfico.
3. Faça a regressão. Tem sentido? Calcule y para x = 95
4. Calcule o erro quadrático médio

```python
import numpy as np
from numpy.random import randn
from numpy.random import seed
from scipy.stats import pearsonr

# Dados
x = np.array([32.50234527, 53.42680403, 61.53035803, 47.47563963, 59.81320787,
              55.14218841, 52.21179669, 39.29956669, 48.10504169, 52.55001444,
              45.41973014, 54.35163488, 44.1640495 , 58.16847072, 56.72720806,
              48.95588857, 44.68719623, 60.29732685, 45.61864377, 38.81681754])
y = np.array([31.70700585, 68.77759598, 62.5623823 , 71.54663223, 87.23092513,
              78.21151827, 79.64197305, 59.17148932, 75.3312423 , 71.30087989,
              55.16567715, 82.47884676, 62.00892325, 75.39287043, 81.43619216,
              60.72360244, 82.89250373, 97.37989686, 48.84715332, 56.87721319])

# Cálculo da correlação de Pearson
corr, _ = pearsonr(x, y)
print(f"Correlação de Pearson: {corr}")
```

```python
import matplotlib.pyplot as plt

# Gráfico de dispersão
plt.scatter(x, y)
plt.title('Gráfico de Dispersão de x e y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# Regressão linear
coef = np.polyfit(x, y, 1)  # Ajuste de 1º grau (linear)
polynomial = np.poly1d(coef)

# Calcular y para x = 95
x_new = 95
y_pred = polynomial(x_new)
print(f"Valor de y para x = {x_new}: {y_pred}")

# Plotar a linha de regressão
plt.scatter(x, y, label='Dados')
plt.plot(x, polynomial(x), color='red', label='Linha de Regressão')
plt.scatter(x_new, y_pred, color='green', label='Predição para x=95')
plt.title('Gráfico de Dispersão com Linha de Regressão')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

Já neste caso, faz sentido, pois a correlação é mais próxima de 1 do que de 0, ou seja, ao mudar o X, o Y é afetado.

```python
# Calcular as previsões
y_pred_all = polynomial(x)

# Erro Quadrático Médio
mse = np.mean((y - y_pred_all) ** 2)
print(f"Erro Quadrático Médio: {mse}")
```
